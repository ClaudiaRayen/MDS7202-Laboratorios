*   **Pregunta 1**: `Git` es un sistema de control de versiones que permite rastrear y gestionar los cambios en proyectos de software y otros archivos a lo largo del tiempo. Entre sus principales caracteristicas permite el seguimiento de las modificaciones, la colaboración entre equipos, la creación de ramas para nuevas características y la fusión de cambios. Git almacena los cambios en repositorios y registra instantáneas de los cambios llamados "commits".
*   **Pregunta 2**:


1.   Funcionalidad: Git es un sistema de control de versiones distribuido que se utiliza para rastrear y gestionar cambios en el código fuente. GitHub, en contraste, es una plataforma web que se basa en Git y amplía sus capacidades para permitir la colaboración en línea y ofrece herramientas adicionales.
2.   Acceso: Git se utiliza principalmente a través de una interfaz de línea de comandos GitHub, por otro lado, proporciona una interfaz gráfica de usuario en línea que facilita el uso de Git y agrega funcionalidades adicionales para la colaboración.
3.  Ubicación:  Git crea repositorios locales para rastrear los cambios locales en un proyecto. GitHub, en cambio, es una plataforma en la nube que aloja repositorios Git en línea, lo que facilita la colaboración en línea y el acceso global a proyectos.


*   **Pregunta 3**: Un `hash` es una función matemática que toma una entrada (ej: archivo, conjunto de datos) y devuelve una cadena alfanumérica única y fija de caracteres, que actúa como una "huella digital" de los datos de entrada, si se genera un dato en los datos de entrada el `hash` generara  una cadena alfanumérica nueva.

Por otro lado, un `commit` es una instantánea de los cambios realizados en un repositorio. Registra una serie de cambios en archivos y carpetas, junto con datos como el autor del cambio, la fecha y hora del commit, y un mensaje descriptivo que explica el propósito del commit. Cada commit en Git se identifica de manera única mediante un hash.

Por ejemplo, si se tuviese un repositorio con un archivo de cualquier tipo y realizace cambios en ese archivo, despues del cambio se debe generar un commit para registrarlo, el cual se identificara con un hash del tipo 85a3f8e7b9d3527f61a591f7d7aecc6e1e3dgge467, el nombre de quien efectuo el cambio en el archivo, el datatime y los cambios realizados:

git commit -m "Cambio en archivo X por [nombre autor] el [fecha y hora]: [descripción de los cambios]"
*   **Pregunta 4**

En un modelo de datos de Git se tienen objetos, que son la base fundamental de Git y almacenan diferentes tipos de datos, como contenido de archivos, árboles de directorios y registros de confirmaciones:

>* Blob: Almacena el contenido de un archivo en un estado específico.
*Tree: Representa la estructura de directorios en un momento dado y contiene referencias a blobs y otros árboles.
*Commit : Almacena la información sobre un conjunto de cambios, incluido un árbol de directorios que representa el estado del proyecto, un mensaje de confirmación y referencias a confirmaciones anteriores.

Cada uno de estos elementos contiene un hash.

Otros elementos son las referencias que son nombres que apuntan a objetos esécíficos:

>* Branches (ramas): Son nombres que apuntan a una confirmación específica. Los branches permiten trabajar en paralelo en diferentes líneas de desarrollo.
*Tags (etiquetas): Son nombres que se asignan a confirmaciones específicas para marcar versiones importantes o hitos en el proyecto.
*HEAD: Es una referencia especial que indica la confirmación actual en la rama activa.

También está el standing area que consiste en el conjunto de archivos modificados que aún no han sido aprobados para formar parte del repositorio ni de su registro de cambios, pero que están a un paso de serlos.

*   **Pregunta 11**; permite guardar temporalmente los cambios no comprometidos en una pila llamada "stash". Esto es útil cuando se necesita cambiar de contexto rápidamente, como cambiar de rama o trabajar en una tarea diferente, sin tener que comprometer los cambios actuales. Los cambios se almacenan en un estado oculto, y luego se pueden aplicar nuevamente en el futuro, manteniendo el historial limpio y organizado.
*   **Pregunta 12**;  `git reset` se utiliza para mover la rama actual a un commit específico y deshacer los cambios posteriores. Puede utilizarse para deshacer cambios locales y reorganizar la historia de confirmaciones. Por otra parte,  `git revert`   crea una nueva confirmación que deshace los cambios de la confirmación especificada. Esto significa que la confirmación original permanece en la historia, pero los cambios se deshacen en una nueva confirmación. Esto es útil cuando se desea mantener un registro de la historia completa, incluso de los cambios revertidos.
