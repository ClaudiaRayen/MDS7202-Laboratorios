### 쯏a, pero esto tiene aplicaciones m치s all치 de guardar informaci칩n en casilleros?

Para el caso de im치genes podemos encontrar m칰ltiples aplicaciones con la manipulaci칩n de los tensores y operando matem치ticamente con ellos. Algunas de las aplicaciones m치s conocidas (y que aplicaremos) son las siguientes:

- **Obtener el negativo de una imagen**: Consiste en obtener el color complementario de una imagen, para esto debemos restar la imagen con la m치xima intensidad que podemos encontrar en una imagen (o sea 255).
    
    
- **Pasar a escala de grises una imagen**: Los valores RGB se convierten a escala de grises mediante la f칩rmula NTSC:

    $$ imagen\_gris = 0.299 * Rojo + 0.587 * Verde + 0.114 * Azul $$
    
    Esta f칩rmula representa la percepci칩n relativa de la persona promedio del brillo de la luz roja, verde y azul.
    
    
- **Mejora de contraste**: Son m칰ltiples las t칠cnicas que nos permiten mejorar el contraste de una imagen, pero, una t칠cnica simple para modificar los contrastes consta en obtener un factor de correcci칩n llamado F en base al contraste deseado (C). Luego, es aplicado en la diferencia entre la imagen y 128. De esta forma obtenemos R, que es la imagen con la mejora de contraste deseada.

    $$ F=259*(C+255)/(255*(259-C)) $$
    $$ R=F*(img-128)+128 $$
  
  
- **Convoluci칩n**: Consiste en recorrer una imagen por cada uno de sus canales utilizando una matriz que lleva por nombre Kernel. El kernel, examinar치 los conjuntos de pixeles que recorre, aplicando una multiplicaci칩n de los valores circundantes ,y sumando todos los valores generados de este producto para generar un nuevo pixel en el tensor de salida.

- Pueden reutilizar el c칩digo implementado en las celdas anteriores para implementar los m칠todos `show` e `info` . Sin embargo, No invoquen directamente esas funciones.
- La idea es que la imagen contenida en la clase sea inmutable, por ende, todos los metodos que modifiquen la imagen contenida en el objeto deberan retornar un nuevo objeto de la clase `Imagen` que contenga la imagen modificada.

````python
class Imagen:
    """Clase contenedora de im치genes"""

    def __init__(self, img):
        if isinstance(img, np.ndarray):
            if img.ndim != 3:
                raise ("Cuidado, el argumento debe ser un arreglo de numpy de solo 3 dimensiones")
            if img.shape[-1] != 3:
                raise (
                    "El argumento debe ser un arreglo de numpy de solo 3 dimensiones "
                    "tal que la 칰ltima dimensi칩n solo tiene 3 canales"
                )
            self.imagen = img
        else:
            raise TypeError(
                "Debes entregar un arreglo de numpy como argumento del constructor de "
                "Imagen"
            )

    def show(self):
        """Muestra la im치gen contenida en el objeto.
        Su funcionalidad debe ser igual a la de la funci칩n mostrar_imagen.
        """

        plt.imshow(self.imagen)
        plt.show()

    def info(self):
        """ Imprime las caracter칤sticas de la imagen cargada: Alto y ancho.
        """

        alto, ancho, _ = self.imagen.shape
        return f"Dimensiones de la imagen: Alto = {alto}, Ancho = {ancho}"

    def __add__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar.

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Sumar y luego que en cada pixel mayor a 255 sea asignado el m치ximo.
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        result = self.imagen + other
        result = np.where(result > 255, 255, result)
        return Imagen(result.astype(int))

    def __radd__(self, other):
        """Operaci칩n conmutativa de __add__.

        Hint: deber칤a llamar a __add__...

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        return self.__add__(other)

    def __sub__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar.

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Restar y luego que en cada pixel mayor a 255 sea asignado el m치ximo.
        # Caso similar para valores menores a 0, donde debera asignar el minimo a esos pixeles
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        result = self.imagen - other
        result = np.where(result > 255, 255, result)
        result = np.where(result < 0, 0, result)
        return Imagen(result.astype(int))

    def __rsub__(self, other):
        """Operaci칩n conmutativa de __sub__.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        return self.__sub__(other)

    def __mul__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar.

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Sumar y luego que en cada pixel mayor a 255 sea asignado el m치ximo y
        # cada valor inferior a 0 debe ser 0.
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        result = self.imagen * other
        result = np.clip(result, 0, 255)
        return Imagen(result.astype(int))

    def __rmul__(self, other):
        """Operaci칩n conmutativa de __mul__.

        Hint: deber칤a llamar a __mul__...

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        return self.__mul__(other)
````

### 2.3 Clase de Procesamiento de Imagenes [2 puntos]

Ahora que comprenden las diferentes dimensiones que componen a una imagen (en la pr치ctica), ahora realizaremos diferentes tareas de procesamiento de im치genes. Para esto, deben crear una clase llamada "`LibImagen`" que cumpla los siguientes requisitos:

- [ ] Pasar una imagen a escala de grises, para esto utilice la ecuaci칩n (1) expuesta en este mismo notebook.
- [ ] Obtener los canales R, G y B de forma individual.
- [ ] Crear una funci칩n que pase a negativo de la imagen alojada.

- [ ] Realizar una convoluci칩n sobre la imagen.

> **Nota 游뉦잺**: Todo m칠todo debe tomar una Imagen y retornar una nueva Imagen.

> **Nota 2**: El tipo de datos del arreglo de la imagen que generen o modifiquen debe ser "int". De lo contrario, puede no visualizarse correctamente.


````python
import cv2

class LibImagen():

    def to_negative(self, img_in):
        """Convierte imagen a negativo.

        Parameters
        ----------
        img_in : Imagen
            Objeto Imagen que contiene imagen a procesar.

        Returns
        -------
        Imagen
            Objeto Imagen con la imagen procesada.
        """
        imagen_negativa = 255 - img_in.imagen
        return Imagen(imagen_negativa)

    def to_gray(self, img_in):
        """
        Transforma una imagen en RGB a la escala de grises.

        Parameters
        ----------
        img_in : Imagen
            Objeto Imagen que contiene una imagen.

        Returns
        -------
        Imagen
            Una que contiene una imagen con 3 canales.
            Los 3 canales deben tener los mismos valores.
        """
        imagen = img_in.imagen  # Acceder a la matriz de p칤xeles dentro del objeto Imagen
        imagen_gris = 0.299 * imagen[:, :, 2] + 0.587 * imagen[:, :, 1] + 0.114 * imagen[:, :, 0]
        imagen_gris = imagen_gris.astype(np.uint8)
        imagen_gris = cv2.cvtColor(imagen_gris, cv2.COLOR_GRAY2BGR)
        return Imagen(imagen_gris)

    def get_channel(self, img_in, channel):
        """Obtiene un canal de un color seteando el resto de los canales en 0.

        Parameters
        ----------
        img_in : Imagen
            Objeto Imagen que contiene una imagen.
        channel : str
            Nombre del canal que ser치 seleccionado. Valores posibles: ('r','g' o 'b').

        Returns
        -------
        Imagen:
            Objeto Imagen que contiene una imagen con 3 canales.
            Solo el canal seleccionado debe tener valores distintos a 0.
        """
        imagen = img_in.imagen  # Acceder a la matriz de p칤xeles dentro del objeto Imagen
        canal_seleccionado = imagen.copy()  # Crear una copia de la matriz
        if channel == 'r':
            canal_seleccionado[:, :, 1] = 0
            canal_seleccionado[:, :, 2] = 0
        elif channel == 'g':
            canal_seleccionado[:, :, 0] = 0
            canal_seleccionado[:, :, 2] = 0
        elif channel == 'b':
            canal_seleccionado[:, :, 0] = 0
            canal_seleccionado[:, :, 1] = 0
        else:
            raise ValueError("El canal debe ser 'r', 'g' o 'b'.")

        return Imagen(canal_seleccionado)

    def set_contrast(self, img_in, C):
        """Mejora el contraste de una imagen.

        Parameters
        ----------
        img_in : Imagen
            Objeto Imagen que contiene una imagen.
        C : float
            Par치metro que define el ajuste de contraste.

        Returns
        -------
        Imagen
            Objeto Imagen que contiene una imagen con 3 canales modificados.
        """
        imagen = img_in.imagen

        F = 259 * (C + 255) / (255 * (259 - C))

        imagen_con_contraste = F * (imagen - 128) + 128

        imagen_con_contraste[imagen_con_contraste > 255] = 255
        imagen_con_contraste[imagen_con_contraste < 0] = 0

        return Imagen(imagen_con_contraste.astype(np.uint8))

    def conv_channel(self, img_in, kernel):
        """
        Aplica una convoluci칩n a cada canal de una imagen utilizando un kernel especificado.

        Parameters
        ----------
        img_in : Imagen
            Objeto Imagen que contiene la imagen de entrada.
        kernel : numpy.ndarray
            Matriz de convoluci칩n que se aplicar치 a cada canal de la imagen.

        Returns
        -------
        Imagen
            Objeto Imagen que contiene una imagen con 3 canales convolucionados.
        """
        img = img_in.imagen
        img_out = []
        for i in range(img.shape[-1]):
            img_channel = convolve2d(img[:, :, i],
                                     kernel,
                                     mode="same",
                                     boundary="symm")
            img_out.append(img_channel)
        new_image = np.stack(img_out, axis=2)
        new_image[new_image>255], new_image[new_image<0] = 255, 0
        return Imagen(new_image.astype(int))
````

Para realizar este ejercicio utilizaremos un m칠todo super b치sico para la eliminaci칩n de fondo. Para esto sigue la siguiente receta:
 1. Pase a escala de grises todas las im치genes secuenciales.
 2. Almacene en una variable la resta de las im치genes consecutivas.

   $$ imagen\_out = imagen(t) - imagen(t+1) $$

 3. Establezca un umbral para eliminar algunos artefactos. Pruebe con al menos 3 umbrales.

$$ imagen\_out = imagen\_out> Umbral $$

**Nota**: No es necesario que construya una clase para esta parte.

````python
import cv2
import numpy as np

def dect_mov(sec_img, umbral=70): #a mayor umbral ser치 m치s oscuro el gift de m치s adelante
    # Lista para almacenar las im치genes procesadas
    mov_imgs = []

    # Recorre las im치genes de la secuencia
    for i in range(len(sec_img) - 1):
        # Convertir las im치genes t y t+1 a escala de grises
        img_t = cv2.cvtColor(sec_img[i], cv2.COLOR_BGR2GRAY)
        img_t1 = cv2.cvtColor(sec_img[i+1], cv2.COLOR_BGR2GRAY)

        # Calcula la diferencia entre las im치genes
        diff = cv2.absdiff(img_t, img_t1)

        # Aplicar el umbral para resaltar los objetos en movimiento
        _, thresh = cv2.threshold(diff, umbral, 255, cv2.THRESH_BINARY)

        # Almacena la imagen procesada en la lista
        mov_imgs.append(thresh)

    return mov_imgs
````
