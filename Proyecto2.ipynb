### Nulos y tratamiento

"Outliers: Se abordarán únicamente los valores negativos. En el caso de 'age' (edad), se aplicará un tratamiento especial a los valores extremos. Respecto a las demás variables, se empleará un RobustScaler para abordarlos más adelante, ya que podrían contener valores atípicos.

Principales acciones de limpieza realizadas:
- Edad ('age'): Se asume que los valores negativos son errores de ingreso, por lo que se convierten a valores positivos. Además, se considera que los valores extremos son también errores de ingreso. Por consiguiente, se obtiene el residuo de la división por 100 para asegurar valores mayores que 0 pero menores que 100.
- Variables con valores atípicos negativos: Al igual que con la edad, los valores negativos se asumen como errores de ingreso y se transforman a valores positivos. Respecto a los valores atípicos positivos o extremos, se corregirán utilizando un escalador.
- Ocupación ('occupation'): Cuando se encuentra el valor '_______', se asume como 'SinInf', lo que quiere decir que no se tiene informción del valor real.
- Comportamiento de pago ('payment_behaviour'): Cuando se encuentra el valor '!@9#%8', también se asume como 'SinInf', con el fin de no perder observaciones.
- Valores nulos: Se utiliza un SimpleImputer con la estrategia de la media para abordar los valores faltantes."

```python
def clean_age(x):
    # Tratamiento de valores negativos y extremos en 'age'
    x = -x if x < 0 else x  # Convertir negativos a positivos
    x = x % 100 if x > 100 else x  # Obtener el residuo de la división por 100 para valores extremos
    return x

def clean_negative(x):
    # Tratamiento de valores negativos en columnas específicas
    return -x if x < 0 else x

def data_cleaner(X):
    # Tratamiento de 'age'
    X['age'] = X['age'].apply(clean_age).astype('Int64')

    # Tratamiento de valores negativos en columnas específicas
    negative_cols = ['num_bank_accounts', 'num_of_loan', 'delay_from_due_date']
    for col in negative_cols:
        X[col] = X[col].apply(clean_negative).astype('Int64')

    # Reemplazo de valores '_______' en 'occupation'
    X['occupation'] = X['occupation'].replace('_______', 'SinInf')

    # Reemplazo de valores '!@9#%8' en 'payment_behaviour'
    X['payment_behaviour'] = X['payment_behaviour'].replace('!@9#%8', 'SinInf')

    # Imputación de valores faltantes usando SimpleImputer para columnas específicas
    null_cols = ['changed_credit_limit', 'num_credit_inquiries', 'amount_invested_monthly', 'monthly_balance']
    imputer = SimpleImputer(strategy='mean')
    X[null_cols] = imputer.fit_transform(X[null_cols])

    return X
```

```python
def Feature_Extractor(X):
    # Copiar el DataFrame original para preservar los datos originales
    processed_df = X.copy()

    # Limpiar y discretizar la variable 'age'
    processed_df['age'] = processed_df['age'].abs()  # Convertir valores negativos a positivos
    processed_df['age'] = processed_df['age'] % 100  # Obtener el residuo de la división por 100 para valores extremos
    processed_df['age_agroup'] = pd.cut(processed_df['age'], bins=[0, 30, 40, 50, float('inf')],
                                        labels=['joven', 'adulto', 'medianaedad', 'adultomayor'], right=False)
    processed_df.drop(columns=['age'], inplace=True)  # Eliminar la columna 'age'

    # Crear la característica 'credit_card_utilization'
    processed_df['credit_card_utilization'] = np.where(processed_df['num_bank_accounts'] != 0,
                                                       processed_df['num_credit_card'] / processed_df['num_bank_accounts'],
                                                       0)  # Evitar división por cero
    processed_df['credit_card_utilization'].fillna(0, inplace=True)

    return processed_df
```

```python

def clean_negative(x):
    # Tratamiento de valores negativos en columnas específicas
    return -x if x < 0 else x

def data_cleaner2(X):
    # Tratamiento de valores negativos en columnas específicas
    negative_cols = ['num_bank_accounts', 'num_of_loan', 'delay_from_due_date']
    for col in negative_cols:
        X[col] = X[col].apply(clean_negative).astype('Int64')

    # Reemplazo de valores '_______' en 'occupation'
    X['occupation'] = X['occupation'].replace('_______', 'SinInf')

    # Reemplazo de valores '!@9#%8' en 'payment_behaviour'
    X['payment_behaviour'] = X['payment_behaviour'].replace('!@9#%8', 'SinInf')

    # Imputación de valores faltantes usando SimpleImputer para columnas específicas
    null_cols = ['changed_credit_limit', 'num_credit_inquiries', 'amount_invested_monthly', 'monthly_balance']
    imputer = SimpleImputer(strategy='mean')
    X[null_cols] = imputer.fit_transform(X[null_cols])

    return X
# Tratamiento de datos
cleaner2 = FunctionTransformer(data_cleaner2)
```

```python
import optuna
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline


def objective_lgbm(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 5000)
    max_depth = trial.suggest_int('max_depth', 2, 10)
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1e-1)

    model = LGBMClassifier(verbose=-1, random_state=seed, n_estimators=n_estimators, max_depth=max_depth, learning_rate=learning_rate)
    pipeline = ImbPipeline(steps=[
        ('extractor', feature_extractor),
        ('cleaner', cleaner2),
        ('preprocessing', ct),
        ('balance', SMOTE()), #Balanceo de clases
        ('model', model)
    ])

    # Entrenamiento y evaluación
    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_test)
    score = precision_score(y_test, preds)

    return score

# Crear y ejecutar el estudio de optimización para LightGBM
study_lgbm = optuna.create_study(direction='maximize')
study_lgbm.optimize(objective_lgbm, timeout = 1*60)

print("Mejores parámetros para LightGBM:", study_lgbm.best_params)
```

```python
def objective_xgb(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 5000)
    max_depth = trial.suggest_int('max_depth', 2, 10)
    learning_rate = trial.suggest_loguniform('learning_rate', 0.005, 0.5)
    min_child_weight = trial.suggest_float('min_child_weight', 1, 10)
    subsample = trial.suggest_float('subsample', 0.5, 1.0)
    colsample_bytree = trial.suggest_float('colsample_bytree', 0.5, 1.0)
    reg_alpha = trial.suggest_float('reg_alpha', 0.0, 1.0)
    reg_lambda = trial.suggest_float('reg_lambda', 0.0, 1.0)

    model = XGBClassifier(
        random_state=seed,
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        min_child_weight=min_child_weight,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        reg_alpha=reg_alpha,
        reg_lambda=reg_lambda,
        n_jobs=-1
    )

    pipeline = ImbPipeline(steps=[
        ('extractor', feature_extractor),
        ('cleaner', cleaner2),
        ('preprocessing', ct),
        ('balance', SMOTE()),  # Balanceo de clases
        ('model', model)
    ])

    # Entrenamiento y evaluación
    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_test)
    score = precision_score(y_test, preds)

    return score

# Crear y ejecutar el estudio de optimización para XGBoost
study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(objective_xgb, timeout=1*60)

print("Mejores parámetros para XGBoost:", study_xgb.best_params)
```

En el caso de los modelos basados en árboles, como LightGBM, se pueden utilizar las importancias de características incorporadas que se basan en la ganancia de información o en la frecuencia con que se divide una característica.
    
    * **Alcances**:
      * Proporciona una vista rápida y fácil de entender de las características que el modelo considera importantes.
      * Puede ser engañoso si las características están correlacionadas entre sí, ya que la importancia puede atribuirse de manera desproporcionada a una de las características correlacionadas.

* **SHAP Values**: Ofrece explicaciones locales alrededor de las predicciones individuales, mostrando cómo cada característica contribuyó a la predicción final.

   *  **Alcances**:
      * Proporcionan explicaciones detalladas para predicciones individuales, lo que es útil para el análisis de casos específicos.
      *Pueden ser más difíciles de interpretar y comunicar a una audiencia no técnica debido a su complejidad.

* **Partial Dependence Plots (PDP)**: Estos gráficos muestran cómo cambia la predicción del modelo cuando se varía una característica específica mientras se mantienen las demás constantes en su valor promedio o según otro criterio.

  *  **Alcances**:
    * Permite visualizar las relaciones entre variables de forma intuitiva, mostrando cómo una característica afecta las predicciones del modelo.

Al combinar la importancia de características, SHAP y PDP, se logra un análisis comprehensivo y robusto de la interpretabilidad del modelo. Las importancias de características brindan una visión general, SHAP ofrece explicaciones detalladas para predicciones individuales, y PDP permite explorar las relaciones entre variables, complementando así la comprensión del comportamiento del modelo en diferentes niveles de detalle.

### Relación de variables

Se creo el siguiente gráfico para visualizar que variables más importantes según SHAP. Las primeras 5 serán las que se visualizarán para ver las relaciones.

```python
shap.summary_plot(shap_values[1], X_processed, feature_names=feature_names)
```

```python
from sklearn.inspection import PartialDependenceDisplay

# Calcular la media de los valores absolutos de SHAP por característica
shap_sum = np.abs(shap_values[1]).mean(axis=0)
top_features = np.argsort(shap_sum)[-5:]  # Índices de las 5 características principales

# Elegir una submuestra aleatoria de 1000 observaciones de X_processed
np.random.seed(seed)  # Para reproducibilidad
subsample_indices = np.random.choice(X_processed.shape[0], 1000, replace=False)
X_subsample = X_processed.iloc[subsample_indices]

# Generar los Partial Dependence Plots para las 5 características más importantes
for feature in top_features:
    fig, ax = plt.subplots(figsize=(8, 4))
    PartialDependenceDisplay.from_estimator(
        best_model, X_subsample, [feature], ax=ax, grid_resolution=50, kind='average'
    )
    ax.set_title(f"Partial Dependence Plot para {feature}", fontsize=16)
    ax.set_xlabel(feature, fontsize=12, rotation=45)
    ax.set_ylabel("Predicted probability change", fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=10)
    plt.show()
```

Recordar que estos gráficos muestran cómo cambia la predicción del modelo cuando se varía una característica específica mientras se mantienen las demás constantes, dado esto se tiene lo siguiente:

* Número de cuentas bancarias (num_bank_accounts): La dependencia parcial disminuye drásticamente para valores bajos y luego se estabiliza. Esto puede indicar que tener alguna cuenta bancaria es importante, pero el número adicional de cuentas no tiene un impacto significativo.

* Retraso desde la fecha de vencimiento (delay_from_due_date): Se observa una fluctuación con un incremento general en la dependencia parcial, que se eleva significativamente después de cierto punto. Esto sugiere que los retrasos más largos podrían estar asociados con un mayor riesgo de incumplimiento.

* Número de tarjetas de crédito (num_credit_card): Hay una disminución significativa en la dependencia parcial a medida que el número de tarjetas de crédito aumenta, lo que podría interpretarse como que tener un número alto de tarjetas se asocia con un menor riesgo, posiblemente debido a un mejor historial de crédito o mayor solvencia.

* Tasa de interés (interest_rate): La dependencia parcial muestra un aumento notable en ciertos rangos de la tasa de interés, seguido de una estabilización. Esto puede indicar que el modelo es sensible a cambios en la tasa de interés hasta cierto punto, tras lo cual los cambios adicionales no afectan tanto la predicción.

* Deuda pendiente (outstanding_debt): Se ve un aumento en la dependencia parcial con el incremento de la deuda pendiente hasta cierto valor, después de lo cual la dependencia se estabiliza. Esto puede reflejar que un aumento en la deuda pendiente hasta cierto nivel incrementa el riesgo de incumplimiento, pero luego el riesgo no aumenta mucho más.
